```python
import pandas as pd
import numpy as np
df=pd.read_csv("Students Social Media Addiction.csv")
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
# 1. Overall Summary for Numerical Data
print("--- 1. Statistical Summary of Numerical Features ---")
print(df.describe())

# 2. Count of Categorical Data
print("\n--- 2. Count of Students by Academic_Level ---")
print(df['Academic_Level'].value_counts())

print("\n--- 3. Count of Students by Most_Used_Platform ---")
print(df['Most_Used_Platform'].value_counts())
      
print("\n--- 4. Count of Students by Gender ---")
print(df['Gender'].value_counts())  
```

```python
# Analyze the average score and sleep grouped by Gender
gender_summary = df.groupby('Gender')[['Addicted_Score', 'Sleep_Hours_Per_Night']].mean()

print("\n--- Average Score and Sleep Hours by Gender ---")
print(gender_summary)

# Analyze the usage distribution by Academic Level
usage_by_academic = df.groupby('Academic_Level')['Avg_Daily_Usage_Hours'].agg(['mean', 'median', 'std'])

print("\n--- Usage Hour Metrics Grouped by Academic_Level ---")
print(usage_by_academic.sort_values(by='mean', ascending=False))
```

```python

# Group by 'Relationship Status' and calculate the mean, median, and count 
# of the 'Addicted_Score' for each status.
relationship_analysis = df.groupby('Relationship_Status')['Addicted_Score'].agg(
    ['mean', 'median', 'count']
)

relationship_analysis_sorted = relationship_analysis.sort_values(by='mean', ascending=False)

print("--- Addiction Score Metrics Grouped by Relationship Status ---")
print(relationship_analysis_sorted)
```

```python
# 1. Pivot Table: Average Addiction Score by Relationship Status and Gender
pivot_table_results = pd.pivot_table(
    data=df,
    values='Addicted_Score',
    index='Relationship_Status',
    columns='Gender',
    aggfunc='mean',
    fill_value='N/A'
)
print("\n--- Average Addiction Score by Relationship_Status and Gender ---")
print(pivot_table_results)

# 2. Correlation Analysis
numerical_cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Addicted_Score']
correlation_with_target = df[numerical_cols].corr()['Addicted_Score'].sort_values(ascending=False)

print("\n--- Correlation of Numerical Factors with Addiction_Score ---")
print(correlation_with_target)
```

```python
import pandas as pd
import numpy as np

df['Avg_Daily_Usage_Hours'] = df['Avg_Daily_Usage_Hours'].replace(['-', 'â€”'], np.nan)
df['Avg_Daily_Usage_Hours '] = pd.to_numeric(df['Avg_Daily_Usage_Hours'], errors='coerce')
df['Avg_Daily_Usage_Hours'].fillna(df['Avg_Daily_Usage_Hours'].mean(), inplace=True)


if 'Student_ID' in df.columns:
    df.drop(columns=['Student_ID'], inplace=True, errors='ignore')

# Encode Binary Feature (Gender)
df['Gender_Encoded'] = df['Gender'].map({'Male': 1, 'Female': 0})

# One-Hot Encode Multi-Category Features (Crucial: Relationship Status)
cols_to_onehot = ['Academic_Level', 'Country', 'Most_Used_Platform', 'Affects_Academic_Performance', 'Relationship_Status']

# THIS CREATES df_processed with the new numeric columns
df_processed = pd.get_dummies(df, columns=cols_to_onehot, drop_first=True)

# Final Cleanup: Drop original text columns
cols_to_drop_final = ['Gender'] 
df_processed.drop(columns=cols_to_drop_final, inplace=True, errors='ignore')

print("Data processing successful. df_processed is ready.")
```

```python
# 1. Identify all new One-Hot Encoded Relationship Status columns
relationship_cols = [col for col in df_processed.columns if 'Relationship Status' in col]

# 2. Calculate the correlation matrix
cols_to_correlate = relationship_cols + ['Addicted_Score']
correlation_matrix = df_processed[cols_to_correlate].corr()

# 3. Extract and sort the correlation values with Addicted_Score
status_correlation = correlation_matrix['Addicted_Score'].drop('Addicted_Score').sort_values(ascending=False)

print("\n--- Correlation of Individual Relationship Status Categories with Addicted Score ---")
print(status_correlation.to_string())
```
