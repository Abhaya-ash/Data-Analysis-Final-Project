```python
import matplotlib.pyplot as plt
import seaborn as sns
x_variable = 'Relationship_Status'
y_variable = 'Addicted_Score'

if x_variable not in df.columns or y_variable not in df.columns:
    print(f"Error: One or more columns ('{x_variable}' or '{y_variable}') not found in the DataFrame.")
else:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[x_variable], y=df[y_variable], palette='viridis')

    plt.title(f'Addiction Score Distribution by {x_variable}', fontsize=16)
    plt.xlabel(x_variable)
    plt.ylabel(y_variable)
    
    plt.grid(axis='y', linestyle='--')
    plt.tight_layout()
    plt.show()
    plt.savefig('relationship_vs_addiction_boxplot.png')
    plt.close()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
x_variable = 'Avg_Daily_Usage_Hours'
y_variable = 'Addicted_Score'

plt.figure(figsize=(8, 6))

sns.scatterplot(x=df[x_variable], y=df[y_variable], hue=df['Gender'], palette='coolwarm')
sns.regplot(x=df[x_variable], y=df[y_variable], scatter=False, color='red')

plt.title(f'Relationship between {x_variable} and {y_variable}', fontsize=14)
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.grid(True, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

X_VAR = 'Sleep_Hours_Per_Night'
Y_VAR = 'Addicted_Score'

plt.figure(figsize=(9, 6))
sns.regplot(
    x=df[X_VAR], 
    y=df[Y_VAR], 
    scatter_kws={'alpha':0.6, 's': 50, 'color': '#1f77b4'}, # Scatter dot styling
    line_kws={'color': '#d62728', 'linewidth': 2} # Regression line styling
)
r_value = df[[X_VAR, Y_VAR]].corr().iloc[0, 1]

plt.title(
    f'Relationship Between Sleep Hours and Addiction Score\n(Correlation r = {r_value:.2f})', 
    fontsize=16, 
    fontweight='bold'
)
plt.xlabel("Sleep Hours Per Night (Mental Health Proxy)", fontsize=12)
plt.ylabel("Addiction Score", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('sleep_vs_addiction_scatter_pure.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

X_VAR = 'Mental_Health_Score'
Y_VAR = 'Addicted_Score'
plt.figure(figsize=(9, 6))
sns.regplot(
    x=df[X_VAR], 
    y=df[Y_VAR], 
    scatter_kws={'alpha':0.6, 's': 50, 'color': '#1f77b4'},
    line_kws={'color': '#d62728', 'linewidth': 2} 
)

r_value = df[[X_VAR, Y_VAR]].corr().iloc[0, 1]

plt.title(
    f'Relationship Between Mental Health and Addiction Score\n(Correlation r = {r_value:.2f})', 
    fontsize=16, 
    fontweight='bold'
)
plt.xlabel("Mental Health Score)", fontsize=12)
plt.ylabel("Addiction Score", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.savefig('health_vs_addiction_scatter_pure.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

status_counts = df['Relationship_Status'].value_counts()
colors = ['#FFC300', '#FF5733', '#C70039', '#900C3F', '#581845'] 

plt.figure(figsize=(8, 8))

# Create the pie chart
plt.pie(
    status_counts.values,             
    labels=status_counts.index,      
    autopct='%1.1f%%',              
    startangle=90,                   #
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5} 
)

plt.title('Distribution of Students by Relationship Status', fontsize=16)
plt.axis('equal') 
plt.savefig('pie_chart_relationship_status.png')
plt.show()
```

```python
platform_counts = df['Most_Used_Platform'].value_counts()

plt.figure(figsize=(9, 9))

# Create the pie chart
plt.pie(
    platform_counts.values,          
    labels=platform_counts.index,    
    autopct='%1.1f%%',               
    startangle=90,                   
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5}
)

plt.title('Distribution of Students by Most Used Platform', fontsize=16)

plt.axis('equal') 
plt.savefig('pie_chart_most_used_platform.png')
plt.show()
```
